## KNN regressor:

A KNN regressor is similar to a KNN classifier , finds the K nearest neighbors and estimates the value of the given
test point based on the values of its neighbours. The main difference between KNN
regression and KNN classification is that KNN classifier returns the label that has
the majority vote in the neighborhood, whilst KNN regressor returns the average
of the neighbors’ values.

```r
# set random seed
set.seed(45)
#split the train data into data and label
train.data <- task1a_train_data[1]
train.label <- task1a_train_data[2]
```

```r
#split the test data into data and label
test.data <- task1a_test_data[1]
test.label <- task1a_test_data[2]
```

```r
#printing first few rows of test data and the dimensions of it
head(test.data)
dim(test.data)
```

```r
#printing first few rows of test label and the dimensions of it
head(test.label)
dim(test.label)
```

```r
#using knn classification algorithm, for each test data point, we will predict the test label based on the nearest train data points
knnregressor <- function(train.data, train.label, test.data, K = 3, distance = 'manhattan'){
    
    #getting the length of train and test data points
    train.len <- nrow(train.data)
    test.len <- nrow(test.data)
    
    #for each test data point, we will calculate the distance of that particular test data point with respect to all training data points, and then select
    #an appropriate portion of it.
    dist <- as.matrix(dist(rbind(test.data, train.data), method = distance))[1:test.len, (test.len+1):(test.len+train.len)]
    
    test_label_predicted <- c()
    
    #we will traverse through via for loop for all the test data points, and for each test data point, the portion of train data points obtained are sorted 
    #based on the distances and the average of those train label points is estimated as the label for test data point.
    for (i in 1:test.len){
        nn <- as.data.frame(sort(dist[i,], index.return = TRUE))[1:K,2]
        test_label_predicted[i] <- mean(train.label[nn,])
    }
    
    return (test_label_predicted)
}
```

```r
#calling knnregressor function where 'k' represents no.of training points to be considered as important for each test data point
knnregressor(train.data, train.label, test.data, K = 3)
```

## Why RMSE is used as the best measure?

RMSE is a quadratic scoring rule that also measures the average magnitude of the error. It’s the square root of the average of squared differences between prediction and actual observation. Since the errors are squared before they are averaged, the RMSE gives a relatively high weight to large errors.

$$RMSE = \sqrt{\frac{1}{n}\Sigma_{i=1}^{n}{\Big({y_i - y}\Big)^2}}$$

```r
#Calculating errors of the train and test data using root mean square root value and storing the obtained values with respect to 'K' where 'K' traverse from 
#1 to 30 in the created data frame.

root_mean_square_error <- data.frame('K'=1:30 , 'Train' = rep(0,30), 'Test' = rep(0,30) )
for (k in 1:30){
    root_mean_square_error[k,'Train'] = sqrt(sum((knnregressor(train.data, train.label, train.data, K = k) - train.label) ** 2)/ nrow(train.data))
    root_mean_square_error[k,'Test'] = sqrt(sum((knnregressor(train.data, train.label, test.data, K = k) - test.label) ** 2)/ nrow(test.data))
}
```

```r
#printing the obtained errors
root_mean_square_error
```

```r
#melting the data frame is done, to have two different values for same k , so that we could plot two different graphs in the same plot.
root_mean_square_error_viz <- melt(root_mean_square_error, id='K') 
names(root_mean_square_error_viz) <- c('K', 'type', 'error')
ggplot(data = root_mean_square_error_viz, aes(x = log(1/K), y = error, color = type)) + geom_line() +
       scale_color_discrete(guide = guide_legend(title = NULL)) + theme_minimal() +
       ggtitle("Missclassification Errors")
```
